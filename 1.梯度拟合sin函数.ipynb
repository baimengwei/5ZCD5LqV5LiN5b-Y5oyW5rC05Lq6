```
import torch
import numpy  as np
import matplotlib.pyplot as plt
import time
```

```
class calculate():
  def __init__(self):
    self.target_function = self.target_function
    self.predict_function = self.predict_function 

    self.dimension = 4
    self.learning_rate = 0.001
    self.weight_array = np.array([np.random.uniform() for i in range(self.dimension)])

  def start_execute(self, count):
    
    # import pdb; pdb.set_trace()
    for i in range(count):
      value_x = np.pi * (np.random.rand()-0.5)
      loss = np.square(self.target_function(value_x) - self.predict_function(value_x))
      value_array = [ np.power(value_x, i) for i in range(self.dimension) ]
      delta_loss = np.dot((self.predict_function(value_x)-self.target_function(value_x)) , value_array) * self.learning_rate
      self.weight_array -= delta_loss
      if i%5000 == 0:
        print('%20s, %20s, %80s'%(loss, self.predict_function(value_x), self.weight_array))

  def predict_function(self, value_x):
    value_array = np.array([np.power(value_x, i) for i in range(self.dimension)])
    return np.dot(self.weight_array , value_array)

  def target_function(self, value_x):
    return np.sin(value_x)


```

```
test_grad = calculate()
test_grad.start_execute(100000)
print('-----------')
print(test_grad.predict_function(np.pi/6))
```
